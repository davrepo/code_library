```python
import numpy as np
import pandas as pd 
```

```python
# extract rainfall in cm as a Numpy array, there is a header
rainfall = pd.read_csv("Seattle2014.csv") 

rainfall.shape # (365, 17) 365 days, 17 features
rainfall.head()
```

```python
rainfall.describe()
```

```python
# extract rainfall in cm as a Numpy array, there is a header
rainfall_cm = (rainfall["PRCP"].values)/100 # /100 to get cm
rainfall_cm.shape # (365,) - get 365 days of PRCP
```

```python
# get number of days with rainfall <= 0.0 cm
np.sum(rainfall_cm == 0.0) # 215 days

# get median value for rainfall, where rainfall > 0.0 cm
np.median(rainfall_cm[rainfall_cm > 0.0]) # 0.495 cm <= rainfall_cm[rainfall_cm > 0.0] is a Numpy array
# verbose way
rainy = rainfall_cm > 0 # this is a mask
np.median(rainfall_cm[rainy]) # 0.495 cm
```
