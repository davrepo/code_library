> Package imports

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols 
```

> Data imports

```
Boston = pd.read_csv("Boston.csv")
```

>  Exercise 1

For the polynomial regression model in Exercise sheet 3 (jupyter notebook), make a 3d plot of the fitted values (predicted medv) against both $$X$$ (lstat) and $$X^2$$. You should find that the relationship is a plane seen as a function of $$(X, X^2)$$, whereas it is a parabola seen as a function of $$X$$ alone.

**Solution**
Let's start by fitting the model again:

```
Xs = Boston[['lstat', 'medv']].sort_values(by='lstat') # So it is easier to make a plot of the function
model = ols("medv ~ lstat + I(lstat**2)", Xs)
lm4 = model.fit()
```

Next, let's plot the model as a function of $$X$$ alone in **2D**:

```
Yh = lm4.get_prediction(Xs).predicted_mean

fig, ax = plt.subplots();
ax.plot(Xs['lstat'], Yh, label='medv ~ lstat + I(lstat**2)', color='r');
ax.scatter(Xs['lstat'], Xs['medv'], alpha=.4);
ax.set_ylabel('Yhat');
ax.set_xlabel('Lstat');
ax.set_title('Plot of model 4', weight='bold');
ax.legend();
```

Looks a nice fit!
